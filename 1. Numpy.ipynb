# Numpy
- It is also called numerical python.
- It is faster than Python lists.
- Uses less memory
- Low level, C optimized backend.
- Support vectorized operation (no need of loop)
- using SIMD (single instruction mulltiple data)
- Contigious memory allocation.

```python
# We can verify it by a performance test. 
```

```python
# add two lists (without using Numpy)
import time

l1 = list(range(1_000_000_0))
l2 = list(range(1_000_000_0))

          
start = time.time()
add = [x+y for x,y in zip (l1, l2)]
end = time.time()

print("Total time taken:",end - start, "seconds")
```

```python
# now using Numpy arange , total time taken is almost 10 times less.
import numpy as np
import time

arr1 = np.array(l1)
arr2 = np.array(l2)

start = time.time()
add = arr1 + arr2
end = time.time()

print("Total time taken:", end-start, "seconds")
```
