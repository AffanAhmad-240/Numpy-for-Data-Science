# Broadcasting in Numpy
- Broadcasting is NumPy's ability to perform operations on arrays of different shapes by automatically "stretching" the smaller array to match the larger one -- without actually copying data.

```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
print(a + 10) 
```

```python
# Scaler 10 is broadcasted across all elements of an array
```

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])   # shape (2, 3)

row    = np.array([10, 20, 30]) # shape (3,)

print(matrix + row)
```

```python
# row is broadcast across each row of matrix.
```

```python
threerowsonecol = np.array([[2],
                      [6],
                      [5]])
onerowfourcol = np.array([1,4,8,7])
result = threerowsonecol + onerowfourcol
```

```python
result
```

```python
col = np.array([[1], [2], [3]])  
row = np.array([10, 20, 30])   

print(col + row)
```

```python
# This creates a full 3×3 matrix from two 1D-ish arrays 
```

### Broadcasting Rules
- NumPy compares shapes from the right dimension by dimension. Two dimensions are compatible if they are either equal or one of them is 1.
