- Numpy arrays are *homogeneous*. It means it can only store elements of the same type. This isdifferent from python lists which store elements of mixed data tyoes.
- Numpy supports various datatypes and understanding them is crucial for optimizing memory usage and performance.    

```python
import numpy as np
```

```python
arr1 = np.array([2,5,8,7])
```

```python
arr1.dtype
```

```python
arr2 = np.array([5.2, 8.9, 2.7, 9.6])
```

```python
arr2.dtype
```

```python
arr3 = np.array([True, False, True])
```

```python
arr3.dtype
```

```python
arr4 = np.array(['aeu', 'v', 'd', 'h'])
```

```python
arr4.dtype
```

```python
# complex datatypes
arr5 = np.array([2+4j, 3+6j, 1+7j], dtype = 'complex64')
```

```python
arr5
```

```python
arr5.nbytes
```

```python
arr6 = np.array(['apple', 'mango'] ,dtype = 'str')
```

```python
arr6.dtype
```

```python
# (Object) Elements of mixed datatypes in one array
# this datatype sacrifices performance. Therefore, we only used it when necessary.
arr7 = np.array([{'abc' : 123}, ['apple', 'b','s'], (1,3,5), {2.9,7.4, 6.2}], dtype = object)
```

```python
arr7
```

```python
type(arr7[3])
```

```python
type(arr7[0])
```
