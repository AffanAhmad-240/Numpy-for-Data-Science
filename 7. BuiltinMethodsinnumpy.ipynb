# Built in methods in Numpy

```python
import numpy as np
```

```python
arr1 = np.array([2,5,9,12,45,89])
```

```python
arr2 = np.array([23,56,89,12,34,12])
```

```python
np.add(arr1, arr2)
```

```python
np.append(3, arr1)
```

```python
np.amax(arr2)
```

```python
np.amin(arr2)
```

```python
np.argmax(arr1)     # return the index of max number
```

```python
np.argmin(arr1)
```

```python
np.array_equal(arr1, arr2)
```

```python
np.asmatrix(arr1)
```

```python
np.cumprod(arr2) #First number multiply with 1 then multipy each number with next number. Result of this multipy with next number and so on.
```

```python
np.cumulative_sum(arr1)
```

```python
np.mean(arr2)
```

```python
np.multiply(arr1, arr2)
```

```python
# Simulating a dataset (5 samples (rows), 3 features (columns))
data = np.array([[10, 20, 30],
                [15, 25, 35],
                 [20,30, 40],
                 [25, 35, 45],
                 [30, 40, 50]])

mean = data.mean(axis = 0)
std = data.std(axis = 0)

normal_data = (data - mean)/std
print(normal_data)
```
