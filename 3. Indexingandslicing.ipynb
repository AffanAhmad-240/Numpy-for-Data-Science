```python
# Indexing and slicing
```

```python
import numpy as np
```

```python
arr1 = np.array([2,4,6,8,9,3,0,78])
```

```python
arr1[0]      #it returns the value present at that index
```

```python
arr1[7]
```

```python
arr1[3:6]
```

```python
arr1[5:]
```

```python
arr1[:8]
```

```python
arr1[::2]
```

```python
arr1[2:7:2]           # start:End:Increment
```

```python
# when we use slicing and indexing in numpy , it only show the view. It don't provide copy of original array. In this way 
# it is memory efficient.
# Let understand with the example
```

```python
arr2 = np.array([3,6,9,12,15])
```

```python
arr2[1:5]
```

```python
arr2[2] = 10       # it directly updates changes in the array we created.
```

```python
arr2
```

```python
# we can also access elements using indexing from 2D array
arr = np.array([[5,6,7],[1,2,3]])
```

```python
arr[0]
```

```python
arr[0][0]
```

```python
arr[1][2]
```

```python
# Fancy indexing
```

```python
arr4 = np.array([2,3,4,5,6,7,8])
```

```python
arr4[[0,2,4]]    # it shows the element present at that index only
```

```python
# Boolean Masking
```

```python
arr5 = np.array([2,5,7,9,1,6])
```

```python
mask = arr5 > 5  # it shows all elements greater than 5
```

```python
print(arr5[mask])
```
